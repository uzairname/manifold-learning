# Autoencoders

```python
import json
from vis import load_model_state_dict, get_outputs
from train_utils.train import get_dataloaders
from datasets.clock import ClockConfig
from models.autoencoders import MLPResnetAutoencoder
import matplotlib.pyplot as plt
import torch
import numpy as n

IMG_SIZE=128
LATENT_DIM=2
BATCH_SIZE=64
dataloader = get_dataloaders(
  data_config=ClockConfig(),
  data_size=2**10,
  img_size=IMG_SIZE,
  batch_size=BATCH_SIZE,
)[0]

type_ = "autoencoder"

fig, ax = plt.subplots(4, 4, figsize=(12, 12))

for i in range(16):
  model = load_model_state_dict(
    model_class = MLPResnetAutoencoder,
    img_size=IMG_SIZE,
    latent_dim=LATENT_DIM,
    name="MLPResnetAutoencoder",
    postfix="d22",
    checkpoint=i
  )
  
  latents = []
  labels1d = []
  for _, label1d, _, latent, _ in get_outputs(type_, model, dataloader):
    latents.append(latent.unsqueeze(0).cpu())
    labels1d.append(label1d.unsqueeze(0).cpu())
    
  latents = torch.cat(latents, dim=0)
  labels1d = torch.cat(labels1d, dim=0)
  
  ax[i//4, i%4].scatter(latents[:, 0], latents[:, 1], c=labels1d, cmap="Spectral", s=1)
  ax[i//4, i%4].set_title(f"Checkpoint {i+1}")
plt.tight_layout()
plt.show()

```
